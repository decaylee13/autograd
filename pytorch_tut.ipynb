```python
import torch 
```

```python
#Step 1: Forward pass
N = 10 #number of elemnts per batch 
D_in = 1 #input dimension
D_out = 1 #output dimension

X = torch.randn(N, D_in)
# print(X)

true_W = torch.tensor([[2.0]])
true_B = torch.tensor(1.0)
y_true = X @ true_W + true_B + torch.randn(N, D_out) * 0.1
# print(y_true)
```

```python
W = torch.randn(D_in, D_out, requires_grad=True)
B = torch.randn(1, requires_grad=True)
```

```python
#forward pass
y_hat = X @ W + B
```

```python
print(y_hat)
print(y_true)
```

```python
#Loss Calculation 
squared_error = (y_hat - y_true) ** 2 
loss = squared_error.mean()

#Gradient calculation
loss.backward() #This function populates the .gradient parameter for all of our tensors with requires_grad = True
```

```python
#Entire training Loop:

#Hyperparameters
epochs, lr = 100, 0.01
#Initialization of parameters 
W = torch.randn(D_in, D_out, requires_grad=True)
b = torch.randn(1, requires_grad=True)

for epoch in range(epochs): 
    if epoch % 10 == 0: 
        print(f"Epoch number: {epoch}, loss: {loss.item()}")
    y_hat = X @ W + b 
    loss = torch.mean(((y_true - y_hat) ** 2))

    loss.backward() 
    with torch.no_grad(): 
        W -= lr * W.grad
        b -= lr * b.grad

    W.grad.zero_()
    b.grad.zero_() 
```
