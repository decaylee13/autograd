```python
import torch 
import numpy as np
```

```python
# Creating tensor for input data 
sample_data = [[1,2,3], [4,5,6]]
x_data = torch.tensor(sample_data)
print(x_data)
```

```python
# Creating tensor from numpy array 
numpy_arr = np.array(sample_data)
tensor_numpy_arr = torch.from_numpy(numpy_arr)
print(tensor_numpy_arr)
```

```python
# Creating tensor from another tensor 
tensor_ones = torch.ones_like(tensor_numpy_arr)
tensor_rand = torch.randn_like(tensor_numpy_arr, dtype=torch.float)
print(tensor_ones)
tensor_rand
```

```python
# Random Tensor Initialization 
shape = (2,3)
ones = torch.ones(shape)
print(ones)
zeroes = torch.zeros(shape)
print(zeroes)
random = torch.rand(shape)
print(random)
```

```python
#Tensor Attributes 
print(tensor_numpy_arr.dtype)
print(tensor_numpy_arr.shape)
print(tensor_numpy_arr.device)
```
